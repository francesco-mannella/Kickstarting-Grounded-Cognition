```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.load("storage/visual_weights_0999.npy", allow_pickle=True)
```

```python
w = d
bw = w.T
```

```python
t = np.arange(10)
grid = np.vstack([x.ravel() for x in np.meshgrid(t, t)]).T
```

```python
hh = []
mm = []
s = 0.5
for x in np.linspace(0, 9, 30):
    hrow = []
    mrow = []
    for y in np.linspace(0, 9, 30):
        point = [x, y] 
        h = (1/(np.sqrt(2*np.pi*s**2)))*np.exp(-0.5*(s**-2)*np.linalg.norm(grid - point, axis=1)**2)
        hrow.append(h.reshape(10, 10))
        o = np.dot(h, bw)
        mrow.append(o.reshape(10, 10, 3))
    hrow = np.hstack(hrow)
    hh.append(hrow)
    mrow = np.hstack(mrow)
    mm.append(mrow)
hh = np.vstack(hh)
mm = np.vstack(mm)

plt.figure(figsize=(14, 7))
plt.subplot(121, aspect="equal")
_ = plt.imshow(hh.transpose(1, 0), aspect="equal")
plt.subplot(122, aspect="equal")
_ = plt.imshow(mm.transpose(1, 0, 2), aspect="equal")
```
